```python
import pandas as pd
```

```python
data=pd.read_csv("mtcars.csv")
data['carb'].nunique()
data
```

```python
data.isnull().sum()
```

# Make feature & target columns

```python
x=data[["mpg","cyl","disp"]].values
y=data["carb"].values
```

# feature scalling using standardtization 

```python
from sklearn.preprocessing import StandardScaler # 
dt=StandardScaler().fit_transform(x)
#dt.fit_transform(x)
```

```python
print(dt.max())
print(dt.min())
```

#  feature scalling using Normalization

```python
from sklearn.preprocessing import MinMaxScaler
dt=MinMaxScaler(feature_range=(1,2)).fit_transform(x)
dt
```

```python
print(dt.max())
print(dt.min())
```

# Splitting & Shuffel data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(dt,y,test_size=0.2,random_state=10)
```

# Import KNN classifier & apply

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(metric='euclidean',p=2)
```

```python
knn.fit(x_train,y_train)
```

```python
pred=knn.predict(x_test)
```
